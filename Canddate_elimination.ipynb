```python
import pandas as pd
import numpy as np 
print("Data")
data = pd.read_csv("../input/data-test2/inp.csv")
print(data)


print("Attributes")
#making an array of all the attributes
d = np.array(data)[:,:-1]
print(d)


out = np.array(data)[:,-1]
n=6
m=4

ans=["phi"]*m

g=[["?"]*n for _ in range(n)]

for i in range(m):
    print("-----------------------STEP {}----------".format(i+1))
    if out[i]=="Yes":
        for j in range(n):
            if ans[j]=="phi":
                ans=d[i]
                break
            else:
                if ans[j]!=d[i][j]:
                    ans[j]="?"
                    g[j][j]="?"
    else:
        for j in range(n):
            if ans[j]!=d[i][j]:
                g[j][j]=ans[j]
            else:
                g[j][j]="?"
    print("ATTRIBUTES",end=" ")
    print(d[i],out[i])
    
    print("SPECIFIC HYPOTHESIS",end=" ")
    print(ans)
        
    print("General Hypothesis",end=" ")
#     for l in g:
#         if l!=["?"]*n:
#             print(l,end=" ")
    print(g)
    

# print(g)
# print(ans)
                    
        

```
